```python
import json

f1s_single = json.load(open('../../_1_one_user_learn_neighbours/scores/f1s_testv_svc.json'))
```

```python
#f1s_general = json.load(open('../../results/f1_scores_general_per_user.json'))
f1s_general_by_split = json.load(open('f1_scores_general_per_user_large.json'))
```

```python
f1s_general_by_split.keys()
```

```python
f1s_general = f1s_general_by_split['u_train']
f1s_general.update(f1s_general_by_split['u_test'])
len(f1s_general)
```

```python
f1s_single = {k: v for (k,v) in f1s_single.items() if k in f1s_general}
```

```python
rows = []
for uid, f1 in f1s_general.items():
    rows.append({
        "uid": uid,
        "f1_general": f1,
        "f1_single": f1s_single[uid],        
    })

import pandas as pd
df = pd.DataFrame(rows)
```

```python
import plotly.graph_objects as go

import numpy as np

x0 = df.f1_single
x1 = df.f1_general

xbins=dict(start=0, end=1.01, size=0.025)
fig = go.Figure()
fig.add_trace(go.Histogram(x=x0, name='single models', bingroup=1, xbins=xbins))
fig.add_trace(go.Histogram(x=x1, name='general model', bingroup=1, xbins=xbins))

# Overlay both histograms
fig.update_layout(barmode='overlay')
# Reduce opacity to see both histograms
fig.update_traces(opacity=0.75)
fig.show()
```
